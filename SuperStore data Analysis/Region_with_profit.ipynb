```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the Excel file into a pandas DataFrame
df = pd.read_csv('superstore.csv')

# Group the data by 'Region' and calculate the total profit for each region
region_profit = df.groupby('Region')['Profit'].sum()
colors = ['blue', 'green', 'orange', 'red']

# Plot the histogram
plt.bar(region_profit.index, region_profit, color=colors)

# Set the x and y labels
plt.xlabel('Region')
plt.ylabel('Profit')

# Add profit values on each bar
for i, (region, profit) in enumerate(region_profit.items()):
    if region == 'West':
        plt.text(i, profit, f'{profit:.2f}', ha='center', va='bottom', color='white')
    else:
        plt.text(i, profit, f'{profit:.2f}', ha='center', va='bottom')

# Find the region with the highest profit
highest_profit_region = region_profit.idxmax()
highest_profit = region_profit.max()

# Annotate the highest profit region on the chart
plt.annotate(f'Highest Profit: {highest_profit}', xy=(highest_profit_region, highest_profit),
             xytext=(highest_profit_region, highest_profit + 500), ha='center',
             arrowprops=dict(arrowstyle='->'))

# Show the chart
plt.show()

# Print the region with the highest profit
print(f'Region with the highest profit: {highest_profit_region}')
```
