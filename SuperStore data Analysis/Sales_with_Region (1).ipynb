```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the Excel file into a pandas DataFrame
df = pd.read_csv('superstore.csv')

# Group the data by region and calculate the sum of sales for each region
region_sales = df.groupby('Region')['Sales'].sum()
colors = ['red', 'orange', 'blue', 'green']
# Plot the bar chart
plt.bar(region_sales.index, region_sales.values,color=colors)
plt.xlabel('Region')
plt.ylabel('Sales')
plt.title('Sales by Region')

# Add labels for the total sale on the left side of the graph
for i, v in enumerate(region_sales.values):
    plt.text(i, v, str(v), ha='right', va='bottom')

plt.show()

# Find the region with the highest sales
highest_sales_region = region_sales.idxmax()
print("Region with the highest sales:", highest_sales_region)
```
