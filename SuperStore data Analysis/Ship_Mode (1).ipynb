```python
import pandas as pd
import matplotlib.pyplot as plt


# Readingthe file
df = pd.read_csv('superstore.csv')

# extract ship mode column
ship_modes = df['Ship Mode']
#count repaetability of each mode
ship_modes_count = ship_modes.value_counts()

#plot pie chart
explode = (0,0,0.1,0)
plt.pie(ship_modes_count, labels=ship_modes_count.index, autopct='%1.1f%%', explode=explode)
plt.axis('equal')
plt.title('Distribution of ship Modes')
plt.legend()
plt.show()
```
