```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('superstore.csv')
df['Discount'] = df['Discount'].apply(lambda x: float(str(x).strip('%')))

average_discount_by_region = df.groupby('Region')['Discount'].mean()

# Define a color palette for the bars
colors = ['blue', 'green', 'orange', 'red']

plt.bar(average_discount_by_region.index, average_discount_by_region.values, color=colors)
plt.xlabel('Region')
plt.ylabel('Average Discount')
plt.title('Average Discount by Region')
plt.legend()

plt.show()

region_with_highest_discount = average_discount_by_region.idxmax()
print("Region with the highest average discount:", region_with_highest_discount)
```
