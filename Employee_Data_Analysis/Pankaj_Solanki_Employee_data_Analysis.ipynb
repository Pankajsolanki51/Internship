###  Principal Component Analysis

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer

# Assuming 'data' is your DataFrame containing the relevant columns
data = data[['EmployeeID', 'EnvironmentSatisfaction', 'JobSatisfaction', 'WorkLifeBalance']]

# Drop any rows with missing values
data = data.dropna()

# Select the columns for PCA
columns_for_pca = ['EnvironmentSatisfaction', 'JobSatisfaction', 'WorkLifeBalance']

# Standardize the data
data_standardized = (data[columns_for_pca] - data[columns_for_pca].mean()) / data[columns_for_pca].std()

# Impute missing values (if any)
imputer = SimpleImputer(strategy='mean')
data_imputed = imputer.fit_transform(data_standardized)

# Instantiate PCA object
pca = PCA()

# Fit PCA model
pca.fit(data_imputed)

# Obtain principal components
principal_components = pca.transform(data_imputed)

# Create a DataFrame to store the results
pca_results = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2', 'PC3'])

# Add EmployeeID to the results DataFrame
pca_results['EmployeeID'] = data['EmployeeID']

# Print explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_
print("Explained Variance Ratio:", explained_variance_ratio)
print()

# Display the results
print("Principal Components:")
print(pca_results)
```
